```python
# Libraries
import numpy as np
from oqc.optimal_quantum_control import OptimalQuantumControl
from qiskit import IBMQ
import matplotlib.pyplot as plt
import pandas as pd

# Load the backend
IBMQ.load_account()
provider = IBMQ.get_provider(group='open')
backend = provider.get_backend('ibmq_armonk')
backend = provider.backend.ibmq_armonk
```

```python
# Define initial parameters
initial_control_params = np.random.random(15)
time_derivative = 1
target_gate = np.array([[0,1], [1, 0]])

# Create the OptimalQuantumControl
control = OptimalQuantumControl(initial_control_params, backend, time_derivative, target_gate)
```

```python
control._backend_amplitude = 1
control._backend_frequency = 1
control._time_derivative = 0.5
```

```python
control.control()
```

```python
y = control._fidelities
x = [i for i in range(len(y))]
```

```python
df = pd.DataFrame({'Iterations':x,'Infidelity':y})
plt.rcParams["figure.figsize"] = (12,8)
plt.xlabel("Iterations")
plt.ylabel("Infidelity")
plt.semilogy(x,y)
plt.savefig('teset.png', format='png', dpi=400)
```
